### Note: mean and std are always defined over a DATASET, not single samples. In this case: per class.

1.15: Assume two 2D normal distributions with different means but equal covariance matrices (i.e. same variance) -> LDA (actually Nearest Mean).

```python
# 1.15: Generate 10 2D data points per class, so 20 data points in total
# The 2 features have different means but the same covariance matrix

# TODO: experiment with diff training set sizes
class_size = 10

# Define the mean for each class
mean_class1 = [-2, 5]
mean_class2 = [8, 3]

# Define the same covariance matrix for both classes (that does not mean all features have the same variance!)
cov = [[5, 2], 
       [3, 3]]

# Generate 10 samples for each class
class1_samples = np.random.multivariate_normal(mean_class1, cov, class_size)
print('Shape of class1_samples: ', class1_samples.shape)  # (10,2)
class2_samples = np.random.multivariate_normal(mean_class2, cov, class_size)
data = np.concatenate((class1_samples, class2_samples), axis=0)

# Verify by computing mean and std -> equal cov matrix also means equal std for both classes
# NOTE: the stds will only be equal for large enough no. of samples!
mean_1 = np.mean(class1_samples, axis=0) 
mean_2 = np.mean(class2_samples, axis=0) 
std_1 = np.std(class1_samples, axis=0) 
std_2 = np.std(class2_samples, axis=0)
# print('Mean1 = ', mean_1)
# print('Mean2 = ', mean_2)
# print('Std1 = ', std_1)
# print('Std2 = ', std_2)

# Create labels: 10 1s and 10 2s starting with 1s
lab = np.concatenate((np.ones(class_size), 2*np.ones(class_size)), axis=0).reshape(2 * class_size, 1)
a = pr.prdataset(data, lab)
# print('Dataset: \n', a.data)
pr.scatterd(a)

# Estimate LDs and QDs (similar to gaussm above) trained on our normally distributed data
ld = pr.ldc(a)
qd = pr.qdc(a)
print('Estimates for LDA: ', ld)
# pr.plotm(ld, nrlevels=5, colors='r', gridsize=500)   # plot the estimates for LDA
print('Estimates for QDA: ', qd)
# pr.plotm(qd, nrlevels=5, colors='b', gridsize=500)   # plot the estimates for QDA

# Plot decision boundaries using plotc
# For multiple classifiers, use plt.legend()
# NOTE: for infinite samples, QDA and LDA will give the same decision boundary (both linear)! 
plotc_ld = pr.plotc(ld, colors='r', gridsize=500)
plotc_qd = pr.plotc(qd, colors='b', gridsize=500)
plt.legend([plotc_ld, plotc_qd], ['LDA', 'QDA'])
```

## 1.6: Density Estimation using Parzen Densities

PRTools can also be used to estimate densities, not just classifiers!

```python
# Create simple dataset having 20 1D objects for each of 2 classes
a = pr.gendats([200, 200], 1, 8)   # TODO: note that train set size leads to lower optimal h!
help(pr.gendats)

# Define width param h for Parzen kernel
# NOTE: the smaller the h, the more complex the model (more overfitting).
# The larger the h, the more simple the model (more underfitting).
h = 0.2
# Define a Parzen density estimator with width h on entire dataset, ignoring labels
a = pr.prdataset(+a)
w = pr.parzenm(a, h)
# Plot the mapping
pr.scatterd(a)
pr.plotm(w, nrlevels=10, colors='r', gridsize=100)
```

### Estimating h using log-likelihood criterion on training set

```python
# Using the same dataset as above, train Parzen density estimator on different h widths
a = pr.gendats([20, 20], 1, 8) 
a = pr.prdataset(+a)                 # Exclude labels
hs = [0.01, 0.05, 0.1, 0.25, 0.5, 1, 1.5, 2, 3, 4, 5] # h's to try
LL = np.zeros(len(hs))               # Array to store log-likelihoods
for i in range(len(hs)):
    parz = pr.parzenm(a, hs[i])         # Train Parzen density estimator
    # We use + to get the estimadated density for each object in the dataset
    LL[i] = np.sum(np.log(+parz(a)))    # Compute log-likelihood of the data

plt.plot(hs, LL, 'o-')               # Plot log-likelihoods as function of h (elbow)

# NOTE: the elbow is at h=0.5, which is the same as the width we used above
# NOTE: we are computing LL on the TRAINING set now! SO THE "optimal" h is not the best h for generalization
```

### Estimating h using log-likelihood criterion on test set

The better the *test* data fits in the probability density model, the higher the LL (i.e. lower error).

This gives us an unbiased error estimate, since the test set is independent of the train set, but same distribution so LL should be maximal in ideal case!

```python
[train_x, test_x] = pr.gendat(a, 0.5) # Split dataset into training and test set
hs = [0.01, 0.05, 0.1, 0.25, 0.5, 1, 1.5, 2, 3, 4, 5] # h's to try
LL_train = np.zeros(len(hs))         # Array to store log-likelihoods for training set
LL_test = np.zeros(len(hs))          # Array to store log-likelihoods for test set
for i in range(len(hs)):
    parz = pr.parzenm(train_x, hs[i])                # Train Parzen density estimator
    LL_train[i] = np.sum(np.log(+parz(train_x)))     # Compute log-likelihood of the training data (WRONG!!)
    LL_test[i] = np.sum(np.log(+parz(test_x)))       # Compute log-likelihood of the test data

# NOTE: overtraining effect!
plt.plot(hs, LL_train, 'o-')        # Plot log-likelihoods as function of h (elbow)
# NOTE: max at h=1
plt.plot(hs, LL_test, 'o-')         # Plot log-likelihoods as function of h (elbow)
```
